```python
import torch
import numpy as np
print("Imports are ready")
```

```python
tensor_a= torch.tensor([1,2,3,4])
tensor_b= torch.tensor([5,6,7,8])
```

```python
# Addition

tensor_a + tensor_b
```

```python
# Addition Long Hand

torch.add(tensor_a, tensor_b)
```

```python
# Subtraction
tensor_b - tensor_a
```

```python
# Sub function
torch.sub(tensor_b,tensor_a)
```

```python
# Multiplication

tensor_a * tensor_b
```

```python
# Mul function
torch.mul(tensor_a,tensor_b)
```

```python
#Division

tensor_b / tensor_a
```

```python
# Div function

torch.div(tensor_b,tensor_a)
```

```python
# Remainder Modulus

tensor_b % tensor_a
```

```python
# Remainder Function

torch.remainder(tensor_b, tensor_a)
```

```python
# Exponenets / power

tensor_a ** tensor_b
```

```python
# Exponents function

torch.pow(tensor_a, tensor_b)
```

```python
# Another way to wright longhand

tensor_add= tensor_a.add(tensor_b)
tensor_sub= tensor_a.sub(tensor_b)
```
