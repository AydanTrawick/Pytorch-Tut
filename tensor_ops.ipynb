```python
import torch
print("Pytorch is ready")
```

```python
my_torch= torch.arange(10)
my_torch
```

```python
## Reshape and view


my_torch = my_torch.reshape(2,5)
my_torch
```

```python
## Reshape if we dont know the number of items we're using

my_torch2 = torch.arange(33)
my_torch2
```

```python
my_torch2 = my_torch2.reshape(-1,3)
my_torch2
```

```python
my_torch3 = torch.arange(20)
my_torch3
```

```python
my_torch4= my_torch3.view(2,-1)
my_torch4
```

```python
# with reshape and view they will update

my_torch5= torch.arange(10)
my_torch5
```

```python
my_torch6= my_torch5.reshape(2,5)
my_torch6
```

```python
my_torch5[1]= 4141
my_torch5[3]= 12
my_torch5
```

```python
my_torch6
```

```python
# slices

my_torch7= torch.arange(10)
my_torch7
```

```python
# Grab a specific item
my_torch7[1]
```

```python
# Grab a slice

my_torch8= my_torch7.reshape(5,2)
my_torch8
```

```python
my_torch8[:,1]
```

```python
# Return a column

my_torch8[:,1:]
```
